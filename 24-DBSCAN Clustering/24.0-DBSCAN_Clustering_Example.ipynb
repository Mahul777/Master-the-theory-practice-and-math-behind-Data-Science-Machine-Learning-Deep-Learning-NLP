```python
# 🔧 Step 1: Import Required Libraries

from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline
# •	DBSCAN: The clustering algorithm we are going to use.
# •	make_moons: Generates 2D data in the shape of interleaving half circles (moon shape).
# •	StandardScaler: Scales the features so they have a mean of 0 and standard deviation of 1.
# •	matplotlib.pyplot: For plotting data points.

# 🌙 Step 2: Create Moon-Shaped Data

X, y = make_moons(n_samples=250, noise=0.05)
# •	n_samples=250: Total number of points in the dataset.
# •	noise=0.05: Adds some random noise/outliers to the data.
# •	This will create two half-moon shapes which are not linearly separable, perfect for DBSCAN.

X
# array([[ 1.63569825e-01, -1.48191610e-01]])
y
# array([0, 1, 1, 0, 1, 1, 1, 1, 1, 1, 1, 0, 1, 1, 0, 0, 0, 0, 0, 1, 0, 1])

# 🎨 Step 3: Visualize the Data

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Original Data with Labels")
plt.show()
# •	This plot shows how the data is shaped.
# •	c=y: Colors the data points based on the actual label (just for visualization).
# •	DBSCAN will try to find these clusters on its own.

# 📏 Step 4: Scale the Features

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Scaling is important because DBSCAN uses distance (like Euclidean), which is affected by feature scale.
# fit_transform() scales the data to standard normal distribution (mean = 0, std = 1).

# 🤖 Step 5: Apply DBSCAN Algorithm

dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan.fit(X_scaled)
# •	eps=0.3: The radius to search for neighboring points.
# •	min_samples=5: Minimum number of points required to form a dense region (core point).
# •	The algorithm automatically finds clusters based on the density of points.


# 🏷️ Step 6: Check Cluster Labels

labels = dbscan.labels_
print(labels)
# •	labels_ gives the cluster number assigned to each point.
# •	-1 indicates a noise point (outlier).
# •	Other numbers like 0, 1 indicate cluster IDs.

# 📊 Step 7: Visualize the Clusters Found by DBSCAN
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
plt.title("DBSCAN Clustered Data")
plt.show()
# •	This plot shows how DBSCAN has grouped the data.
# •	It should show two moon-shaped clusters, with noise points possibly in a different color.




```
