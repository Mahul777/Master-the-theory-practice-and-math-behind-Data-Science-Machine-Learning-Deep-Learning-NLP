```python
# â€¢	A function is a block of code that does a specific task.
# â€¢	Benefits of using functions:
# o	Makes your code more organized.
# o	Allows you to reuse code without rewriting it.
# o	Improves readability of your code.
# â€¢	You can call a function any time you need it.

# ðŸ”§ Function Syntax in Python
def greet(name):
    """Returns a personalized greeting message."""
    message = "Hello, " + name + "!"
    return message
greet("Apoorv") #'Hello, Apoorv!'


# ðŸ”¹ Why Do We Need Functions?
# Functions help us:

# Avoid repeating code

# Organize our program into reusable blocks

# Make code easier to read and maintain

#ðŸ§ª Example: Checking if a Number is Even or Odd
def even_odd(num):
    if(num%2 ==0):
        print("Even number")
    else:
        print("Odd Number")
even_odd(24) #Even number
even_odd(33) #Odd Number

# ðŸ”¹ Function with Multiple Parameters
# Functions can take more than one parameter to perform more complex tasks.

# You can pass as many parameters as needed, separated by commas.

def add(a,b):
    return a+b

result = add(2,3)
print(result) #5

# ðŸ”¹ Default Parameters
# You can assign a default value to a parameter in a function.

# This allows the function to work even if the argument is not provided.

def greet(name="daale Gopal"):
    print(f"Hello, {name} what is going on")

greet()
greet("Mohit")

# âœ… Positional Arguments (*args)
# Use an asterisk * to allow a function to accept any number of positional arguments.

# These arguments are collected into a tuple.
def func(*number):
    for num in number:
        print(num)
func(1,2,3,4,5,6) # 1 2 3 4 5 6

# âœ… Keyword Arguments (**kwargs)
# Use double asterisks ** to accept any number of keyword arguments.

# These arguments are collected into a dictionary.

def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}") # name: Krish age: 30 city: Bangalore
print_info(name="Krish", age=30, city="Bangalore")




```
