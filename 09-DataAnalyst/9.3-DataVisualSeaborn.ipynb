🔹 Introduction to Seaborn
•	📌 Seaborn is a Python data visualization library.
•	🧱 Built on top of Matplotlib.
•	✨ Makes it easy to create beautiful, informative statistical plots.
•	🧰 Requires less code for complex visualizations.
•	🎯 This guide covers basic to advanced Seaborn plots

```python
import seaborn as sns
import matplotlib.pyplot as plt  # ✅ Import matplotlib for plotting

# Load the tips dataset
tips = sns.load_dataset('tips')
print(tips.head())

# Create a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()

# create a line plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
# 4️⃣ Box Plot
# •	📦 Shows data distribution and outliers.
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box plot of Total Bill by Day")
plt.show()
```

```python
# 5️⃣ Violin Plot
# •	🎻 Combination of box plot + KDE (density)
sns.violinplot(x='day', y='total_bill', data=tips)
# 📌 Helps visualize the density and spread of values.
```

```python
# 6️⃣ Histogram with KDE
# •	📉 Shows frequency of values + smooth distribution curve
sns.histplot(tips['total_bill'], bins=10, kde=True)
# •	Remove KDE: kde=False
```

```python
# 7️⃣ KDE Plot (Only Curve)
# •	🌊 Visualizes density with just the KDE curve
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
# 8️⃣ Pair Plot
# •	🔗 Shows pairwise scatter plots and histograms for numeric features
sns.pairplot(tips)
# 📌 Great for EDA (Exploratory Data Analysis)
```

```python
# 9️⃣ Heatmap (Correlation Matrix)
# •	🔥 Highlights relationships between numerical variables
corr = tips[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
# 📌 Correlation values:
# •	+1 → strong positive
# •	-1 → strong negative
# •	0 → no correlation

# Bokeh Visualizations 
# Bokeh is a powerful library for creating interactive visualizations in Python.
# It allows you to create web-based visualizations that can be easily shared and embedded.
# Example: Creating a simple Bokeh plot
%pip install bokeh
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
# Create a simple Bokeh plot
p = figure(title="Simple Bokeh Plot", x_axis_label='x', y_axis_label='y')
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], legend_label="Line", line_width=2)
show(p)
```

```python
# 📦 Example: Using 'tips' dataset instead of 'sales_data'
# 🧮 Bar Plot – Total Bill by Day
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum)
plt.title('Total Bill by Day (Sum)')
plt.show()
# •	🛍️ Shows total bill by day.

# 🌍 Bar Plot – Total Bill by Time (Lunch/Dinner)
sns.barplot(x='time', y='total_bill', data=tips, estimator=sum)
plt.title('Total Bill by Time (Sum)')
plt.show()
# •	📌 Helps identify high-performing times (Lunch vs Dinner)
```
