```python
import seaborn as sns
import matplotlib.pyplot as plt  # âœ… Import matplotlib for plotting

# Load the tips dataset
tips = sns.load_dataset('tips')
print(tips.head())

# Create a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()

# create a line plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
# 4ï¸âƒ£ Box Plot
# â€¢	ðŸ“¦ Shows data distribution and outliers.
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box plot of Total Bill by Day")
plt.show()
```

```python
# 5ï¸âƒ£ Violin Plot
# â€¢	ðŸŽ» Combination of box plot + KDE (density)
sns.violinplot(x='day', y='total_bill', data=tips)
# ðŸ“Œ Helps visualize the density and spread of values.
```

```python
# 6ï¸âƒ£ Histogram with KDE
# â€¢	ðŸ“‰ Shows frequency of values + smooth distribution curve
sns.histplot(tips['total_bill'], bins=10, kde=True)
# â€¢	Remove KDE: kde=False
```

```python
# 7ï¸âƒ£ KDE Plot (Only Curve)
# â€¢	ðŸŒŠ Visualizes density with just the KDE curve
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
# 8ï¸âƒ£ Pair Plot
# â€¢	ðŸ”— Shows pairwise scatter plots and histograms for numeric features
sns.pairplot(tips)
# ðŸ“Œ Great for EDA (Exploratory Data Analysis)
```

```python
# 9ï¸âƒ£ Heatmap (Correlation Matrix)
# â€¢	ðŸ”¥ Highlights relationships between numerical variables
corr = tips[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
# ðŸ“Œ Correlation values:
# â€¢	+1 â†’ strong positive
# â€¢	-1 â†’ strong negative
# â€¢	0 â†’ no correlation

# Bokeh Visualizations 
# Bokeh is a powerful library for creating interactive visualizations in Python.
# It allows you to create web-based visualizations that can be easily shared and embedded.
# Example: Creating a simple Bokeh plot
%pip install bokeh
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
# Create a simple Bokeh plot
p = figure(title="Simple Bokeh Plot", x_axis_label='x', y_axis_label='y')
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], legend_label="Line", line_width=2)
show(p)
```

```python
# ðŸ“¦ Example: Using 'tips' dataset instead of 'sales_data'
# ðŸ§® Bar Plot â€“ Total Bill by Day
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum)
plt.title('Total Bill by Day (Sum)')
plt.show()
# â€¢	ðŸ›ï¸ Shows total bill by day.

# ðŸŒ Bar Plot â€“ Total Bill by Time (Lunch/Dinner)
sns.barplot(x='time', y='total_bill', data=tips, estimator=sum)
plt.title('Total Bill by Time (Sum)')
plt.show()
# â€¢	ðŸ“Œ Helps identify high-performing times (Lunch vs Dinner)
```
