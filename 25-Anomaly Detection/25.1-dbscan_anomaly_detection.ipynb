```python
# ✅ Step 1: 📦 Import Necessary Libraries
 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
# 🔹 numpy — for numerical operations
# 🔹 matplotlib.pyplot — to plot the data
# 🔹 make_circles — generate synthetic circular data
# 🔹 DBSCAN — the clustering algorithm

# ✅ Step 2: 🎯 Generate Circular Dataset with Noise
 
X, y = make_circles(n_samples=750, factor=0.5, noise=0.05, random_state=42)
X
# array([[-0.31809126,  0.45037727]])

# 🌀 Creates 750 points in two concentric circles
# factor controls the relative size of the inner circle compared to the outer circle.
# 🧪 Adds 5% noise to simulate real-world imperfections
# 🎲 random_state=42 ensures reproducibility

# ✅ Step 3: ⚙️ Apply DBSCAN Algorithm
 
dbscan = DBSCAN(eps=0.1, min_samples=4)
labels = dbscan.fit_predict(X)
labels
# array([ 0,  1,  2,  2,  2,  0,  2,  2,  0,  0,  2,  2,  2,  0,  2,  2,  1])
# 📏 eps=0.1 → radius for neighbors
# 🔢 min_samples=4 → at least 4 points = core point
# 🧠 fit_predict assigns a cluster label to each point
# 0, 1, 2... → cluster IDs
# -1 → ❌ outlier

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=10)
plt.title("DBSCAN Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()


```
