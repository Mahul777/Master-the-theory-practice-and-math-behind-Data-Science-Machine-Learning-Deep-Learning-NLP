```python
#Tuple
# Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
empty_tuple =();
print(type(empty_tuple)) #<class 'tuple'>

# Tuple Decleration with tuple function 
my_tuple = tuple([1,2,3,4])
print(my_tuple) # (1, 2, 3, 4)

# Adding Elements to a Tuple
# Tuples are immutable, so you cannot add elements directly.
# However, you can convert the tuple to a list, add elements to the list, and then convert it back to a tuple.
my_tuple = (1,2,3)
temp_list = list(my_tuple) # Convert tuple to list
temp_list.append(4) # Add element to list
my_tuple = tuple(temp_list) # Convert list back to tuple
print(my_tuple) # (1, 2, 3, 4)

# Mixing Data Types in a Tuple
mixed_tuple = tuple([1,2,"Hello",2.41])
print(mixed_tuple) # (1, 2, 'Hello', 2.41)

# üìå Accessing Tuple Elements
# ‚Ä¢	Use indexing to access elements:
# ‚Ä¢	numbers[0]       # First element
# ‚Ä¢	numbers[-1]      # Last element
# ‚Ä¢	Use slicing to get a range of elements:
# ‚Ä¢	numbers[0:4]         # Elements from index 0 to 3
# ‚Ä¢	numbers[::-1]        # Reversed tuple
print(mixed_tuple) # (1, 2, 'Hello', 2.41)
print(mixed_tuple[0]) # 1
print(mixed_tuple[-1]) # 2.41
print(mixed_tuple[0:5])  # (1, 2, 'Hello', 2.41)
print(mixed_tuple[::-1]) # (2.41, 'Hello', 2, 1)

# üìå Tuple Operations
# i)‚ûï Concatenation
tuplr1 =(1,2,3)
tuplr2 =(4,5,6)
tuple3= tuplr1 + tuplr2
print(tuple3) # (1, 2, 3, 4, 5, 6)

# ii)‚ûï Repetition
tuplr1 =(1,2,3) 
print(tuplr1*3) # (1, 2, 3, 1, 2, 3, 1, 2, 3)
# tuplr1 * 3 means repeat the contents of t 3 times

# iii)‚ûï Membership Test
tuple1=(1,2,3,4)
print(1 in tuple1)  # True
print(6 not in tuple1 ) #True
tuple1 =(2,2,2)
print(tuple1)
#tuple1[0] = 5 
#print(tuple1) #tuple' object does not support item assignment

# üìå Tuple Methods
# 1. üî¢ count()
t = (1,2,3,2,2)
print(t.count(2)) #3

# 2. üîç index()
print(t.index(2))   #1

# | Operation       | Example                | Description               |
# | --------------- | ---------------------- | ------------------------- |
# | `len()`         | `len(t)`               | Get number of items       |
# | `min()`         | `min(t)`               | Minimum value in tuple    |
# | `max()`         | `max(t)`               | Maximum value in tuple    |
# | `sum()`         | `sum(t)`               | Sum of numeric elements   |
# | `in` / `not in` | `2 in t`, `5 not in t` | Membership test           |
# | `tuple()`       | `tuple([1, 2, 3])`     | Convert iterable to tuple |

print(t) #(1, 2, 3, 2, 2)
print(len(t)) #5
print(min(t)) #1
print(max(t)) #3 is largest element
print(sum(t)) #10

# üì¶ Packing and Unpacking Tuples

# 1. Packing Tuples
# Packing a tuple means putting multiple values into a tuple and assigning it to a variable.
t =(1,2,3,4)
print(t) #(1, 2, 3, 4)

# 2. Unpacking Tuples
# Unpacking means extracting values from a tuple and assigning them to individual variables.
a,v,c,d =t
print(a) #1
print(v) #2
print(c) #3
print(d) #4

# 3. Extended Unpacking
# Sometimes you don't know exactly how many elements a tuple has, and you want to capture some values while ignoring others. Python allows for extended unpacking using the * operator.

t = (1, 2, 3, 4, 5)

# Unpack the first value and capture the rest in a list
first, *rest = t
print(first)  # Output: 1
print(rest)   # Output: [2, 3, 4, 5]

# 4. Unpacking with Discards
# If you don‚Äôt need some of the values during unpacking, you can use an underscore (_) as a placeholder.
a,_,_,_,b =t
print(a) # 1
print(b) # 5
```

```python
# üß± Nested Tuples and Lists
nested_list = [[1,2,3],[6,7,8],[12,"Hello",3.14]]
# Accessing elements
nested_list[0] # [1, 2, 3]
nested_list[0][1] #2
# ‚Ä¢	Slicing
nested_list[1][0:3] #[6, 7, 8]

# ‚Ä¢	A list can have tuples too:
my_list = [[1,2,3],(1,"Hello",1.2)]
print(my_list[1][1]) #Hello

# Nested Tuples and Looping
nested = ((1, 2), (3, 4), (5, 6))

for pair in nested:
    print(pair) #(1, 2) (3, 4) (5, 6)
```
