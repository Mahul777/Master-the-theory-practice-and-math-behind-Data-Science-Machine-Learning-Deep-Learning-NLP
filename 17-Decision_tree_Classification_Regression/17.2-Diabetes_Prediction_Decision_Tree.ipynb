```python
# ✅ Step 1: Import the Dataset
# Use sklearn.datasets to load the diabetes dataset.
# This dataset is already built into sklearn.

from sklearn.datasets import load_diabetes

# Load the dataset
data = load_diabetes()
# data now contains:
# data.data: Independent features (X)
# data.target: Dependent variable (Y)
# data.DESCR: Description of the dataset

# {
#     'data': array([[ 0.03807591,  0.05068012,  0.06169621, ..., -0.00259226,
#           0.01990749, -0.01764613],])
#     'target': array([151.,  75., 141., 206., 135.,  97., 138.,  63., 110., 310., 101.])
#      'frame': None,
#      'DESCR': '.. _diabetes_dataset:\n\nDiabetes dataset,
#      'feature_names': ['age',
#   'sex',
#   'bmi',
#   'bp',
#   's1',
#   's2',
#   's3',
#   's4',
#   's5',
#   's6'],
#  'data_filename': 'diabetes_data_raw.csv.gz',
#  'target_filename': 'diabetes_target.csv.gz',
#  'data_module': 'sklearn.datasets.data'}

# ✅ Step 2: Explore the Dataset
# Check the description to understand features.


print(data.DESCR)
# There are 10 features (independent variables):
# age, sex, bmi, bp, s1, s2, s3, s4, s5, s6

# Target: A numeric value indicating disease progression after 1 year (used for regression)
# :Number of Instances: 442
# :Attribute Information:
    # - age     age in years
    # - sex
    # - bmi     body mass index
    # - bp      average blood pressure
    # - s1      tc, total serum cholesterol
    # - s2      ldl, low-density lipoproteins
    # - s3      hdl, high-density lipoproteins
    # - s4      tch, total cholesterol / HDL
    # - s5      ltg, possibly log of serum triglycerides level
    # - s6      glu, blood sugar level

# ✅ Step 3: Create a DataFrame
# Convert the dataset into a pandas DataFrame for easier handling.


import pandas as pd

# Create DataFrame
df_diabetes = pd.DataFrame(data.data, columns=data.feature_names)
df_diabetes['target'] = data.target  # Add the target column

# View the first few rows
df_diabetes.head()
# | Index | age       | sex       | bmi       | bp        | s1        | s2        | s3        | s4        | s5        | s6        | target |
# | ----- | --------- | --------- | --------- | --------- | --------- | --------- | --------- | --------- | --------- | --------- | ------ |
# | 0     | 0.038076  | 0.050680  | 0.061696  | 0.021872  | -0.044223 | -0.034821 | -0.043401 | -0.002592 | 0.019907  | -0.017646 | 151.0  |
# | 1     | -0.001882 | -0.044642 | -0.051474 | -0.026328 | -0.008449 | -0.019163 | 0.074412  | -0.039493 | -0.068332 | -0.092204 | 75.0   |
# | 2     | 0.085299  | 0.050680  | 0.044451  | -0.005670 | -0.045599 | -0.034194 | -0.032356 | -0.002592 | 0.002861  | -0.025930 | 141.0  |
# | 3     | -0.089063 | -0.044642 | -0.011595 | -0.036656 | 0.012191  | 0.024991  | -0.036038 | 0.034309  | 0.022688  | -0.009362 | 206.0  |
# | 4     | 0.005383  | -0.044642 | -0.036385 | 0.021872  | 0.003935  | 0.015596  | 0.008142  | -0.002592 | -0.031988 | -0.046641 | 135.0  |


# ✅ Step 4: Split into X and Y
# Separate independent (X) and dependent (Y) variables.

X = df_diabetes.drop('target', axis=1)  # Features
y = df_diabetes['target']               # Target
X.head()
# | Index | age       | sex       | bmi       | bp        | s1        | s2        | s3        | s4        | s5        | s6        |
# | ----- | --------- | --------- | --------- | --------- | --------- | --------- | --------- | --------- | --------- | --------- |
# | 0     | 0.038076  | 0.050680  | 0.061696  | 0.021872  | -0.044223 | -0.034821 | -0.043401 | -0.002592 | 0.019907  | -0.017646 |
# | 1     | -0.001882 | -0.044642 | -0.051474 | -0.026328 | -0.008449 | -0.019163 | 0.074412  | -0.039493 | -0.068332 | -0.092204 |
# | 2     | 0.085299  | 0.050680  | 0.044451  | -0.005670 | -0.045599 | -0.034194 | -0.032356 | -0.002592 | 0.002861  | -0.025930 |
# | 3     | -0.089063 | -0.044642 | -0.011595 | -0.036656 | 0.012191  | 0.024991  | -0.036038 | 0.034309  | 0.022688  | -0.009362 |
# | 4     | 0.005383  | -0.044642 | -0.036385 | 0.021872  | 0.003935  | 0.015596  | 0.008142  | -0.002592 | -0.031988 | -0.046641 |
y.head()
# 0    151.0
# 1     75.0
# 2    141.0
# 3    206.0
# 4    135.0
# Name: target, dtype: float64

# ✅ Step 5: Train-Test Split
# Split the data into training and testing parts.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

# ✅ Step 6: Check Correlation (Optional)
# This helps understand how features relate to each other.


import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(X_train.corr(), annot=True, cmap='coolwarm')
plt.show()


# ✅ Step 7: Train Decision Tree Regressor

from sklearn.tree import DecisionTreeRegressor

# Initialize and train the model
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)

# ✅ Step 8: Hyperparameter Tuning (GridSearchCV)
# Try different combinations of parameters using GridSearchCV.

from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter': ['best', 'random'],
    'max_depth': [2, 4, 6, 10, 15],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Create model
regressor = DecisionTreeRegressor()

# Grid Search with 5-fold cross-validation
grid = GridSearchCV(regressor, param_grid, cv=5, scoring='neg_mean_squared_error')
grid.fit(X_train, y_train)

# ✅ Step 9: Best Parameters

print("Best Parameters:", grid.best_params_)

# ✅ Step 10: Evaluate the Model
# Make predictions and check accuracy using regression metrics.

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Predict on test set
y_pred = grid.predict(X_test)

# Metrics
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))

# ✅ Step 11: Visualize the Decision Tree

from sklearn.tree import plot_tree

# Use best model
best_model = grid.best_estimator_

# Plot the tree
plt.figure(figsize=(50,50))
plot_tree(best_model, filled=True, feature_names=data.feature_names)
plt.show()

# Best Parameters: {'criterion': 'squared_error', 'max_depth': 6, 'max_features': 'sqrt', 'splitter': 'random'}
# R2 Score: 0.07580230875802918
# MAE: 64.6258538871321
# MSE: 5848.019016966027
```
